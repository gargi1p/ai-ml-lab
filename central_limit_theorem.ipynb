```python
#the center limit theorm
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#1. generate a population with a non normal uniform distribution
population_size=100
population=np.random.uniform(0,10,population_size)

#plot the population distribution
plt.figure(figsize=(12,5))
sns.histplot(population,kde=True,bins=30,color='skyblue')
plt.title('Population Distribution(Uniform)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

#2. set parameters for the central limit theorem demonstration
sample_size=30  #size of each sample
num_samples=100 #Num of sample to draw

#3. draw multiple sample and calculate their means
sample_means=[]
for _ in range(num_samples):
    sample=np.random.choice(population,sample_size)
    sample_means.append(np.mean(sample))

    #4.plot the distribution of sample means
    plt.figure(figsize=(12,5))
    sns.histplot(sample_means,kde=True,bins=30,color='salmon')
    plt.title('Distribution of Sample Means (Sample Size = (sample_size))')
    plt.xlabel('Sample Means')
    plt.ylabel('Frequency')
    plt.show()
```
